```python
# 1.1. Скачать следующие данные: kc-house-data и laptop_price
import pandas as pd
houses = pd.read_csv('data/asses/kc_house_data.csv', sep = ',')
houses.head()
```

```python
# 1.2. Изучите стоимости недвижимости
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(houses['price'], bins=50, color='skyblue', edgecolor='black')
plt.title('Распределение стоимости недвижимости')
plt.xlabel('Цена')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

```python
# 1.3. Изучите распределение квадратуры жилой
plt.figure(figsize=(10, 6))
plt.hist(houses['sqft_living'], bins=50, color='lightgreen', edgecolor='black')
plt.title('Распределение квадратуры жилой площади')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

```python
# 1.4. Изучите распределение года постройки
plt.figure(figsize=(10, 6))
plt.hist(houses['yr_built'], bins=50, color='lightcoral', edgecolor='black')
plt.title('Распределение года постройки')
plt.xlabel('Год постройки')
plt.ylabel('Частота')
plt.grid(True)
plt.show()
```

```python
# Выводы: дешевых маленьких домов больше, дорогих и больших домов - меньше. Тренд количества зданий имеет положительную тенденцию по годам.
```

```python
# Условие 2: 2 задача
# 2.1. Изучите распределение домов от наличия вида на набережную
# Постройте график
# Сделайте выводы
# 2.2. Изучите распределение этажей домов
# 2.3. Изучите распределение состояния домов
```

```python
import seaborn as sns

# Построение графика распределения домов от наличия вида на набережную
plt.figure(figsize=(8, 6))
sns.countplot(x='waterfront', data=houses, hue='waterfront', palette='viridis', legend=False)
plt.title('Распределение домов от наличия вида на набережную')
plt.xlabel('Вид на набережную')
plt.ylabel('Количество домов')
plt.xticks([0, 1], ['Нет', 'Да'])
plt.grid(True)
plt.show()
```

```python
# Изучение распределения этажей домов
plt.figure(figsize=(8, 6))
sns.countplot(x='floors', data=houses, hue='floors', palette='plasma', legend=False)
plt.title('Распределение этажей домов')
plt.xlabel('Количество этажей')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

```python
# Изучение распределения состояния домов
plt.figure(figsize=(8, 6))
sns.countplot(x='condition', data=houses, hue='condition', palette='magma', legend=False)
plt.title('Распределение состояния домов')
plt.xlabel('Состояние')
plt.ylabel('Количество домов')
plt.grid(True)
plt.show()
```

```python
# Выводы: количество домов с видом на набережную очень мало, большинство домов имеют 1-2 этажей,большинство домов в состоянии 3 из 5.
```

```python
# Условие 3: 3 задача
# Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
# Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.
```

```python
# 1. Scatter Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=houses, alpha=0.5)
plt.title('Зависимость цены от общей площади дома')
plt.xlabel('Площадь дома, кв. футов')
plt.ylabel('Цена, $')
plt.grid(True)
plt.show()
```

Видим с увеличением общей площади дома цена также увеличивается.

```python
# 2. Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=houses)
plt.title('Зависимость цены от количества спален')
plt.xlabel('Количество спален')
plt.ylabel('Цена, $')
plt.grid(True)
plt.show()
```

Цена недвижимости имеет тенденцию к увеличению с увеличением количества спален, но в некоторых вариантах (особенно 2-6 спален) много выбросов (вероятно это самые популярные варианты, с хорошим ремонтом и/или локацией)

```python
# 3. Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=houses, x='price', bins=30, kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

Гистограмма показывает, что большинство цен на недвижимость сосредоточены в нижней и средней части диапазона, с небольшим количеством дорогих объектов. Это подтверждает наличие выбросов в данных.

```python
# Графики средней цены по категориям:
plt.figure(figsize=(12, 6))
sns.barplot(data=houses, x='grade', y='price')
plt.title('Average Price by Grade')
plt.xlabel('Grade')
plt.ylabel('Average Price')
plt.show()
```

```python
Чем выше оценка, тем выше средняя цена
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_columns = houses.select_dtypes(include=['float64', 'int64'])

# Compute correlation matrix
corr_matrix = numeric_columns.corr()

# Create heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Housing Features')
plt.show()

```
