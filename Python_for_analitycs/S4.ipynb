```python
import pandas as pd
df = pd.read_csv("data/S4/laptops_10.csv")
df.head()
```

1.1 Изучить количество памяти с помощью matplotlib
    - постройте график
    - назовите график
    - назвать оси
    - сделайте выводы

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))

plt.hist(df['Memory_Amount'], color="lightgray", edgecolor="black")
plt.title('Распределение памяти')
plt.xlabel('GB')
plt.ylabel('Count');
```

```python
plt.figure(figsize=(10,4))

plt.hist(df['Price_euros'], color="lightgreen", edgecolor="black")
plt.title('Распределение euro')
plt.xlabel('euro')
plt.ylabel('Count');
```

2.  Изучите распределение типов носителя

```python
data = df['Memory_Type'].value_counts()
data
```

```python
names = data.index
values = data.values

plt.pie(values, labels = names, autopct='%.1f%%');
```
