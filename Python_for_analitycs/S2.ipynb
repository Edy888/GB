```python
# Как много ультрабуков с 8гб в процентном соотношении
ultra = df[df['TypeName'] == 'Ultrabook'].shape[0]
ultra8 = df[(df['TypeName'] == 'Ultrabook') & (df['Ram'] == '8GB')].shape[0] #проверка размерности матрицы
ultra8 / ultra * 100
```

```python
# ноут с 8 или 16 до 500 евро
df[
    (df['OpSys'] == 'Windows 10') &
    (df['Price_euros'] < 500) &
    (df['Ram'].isin(['8GB', '16GB']))
].shape[0]
```
